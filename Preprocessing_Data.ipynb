## Preprocessing Data



```python
import pandas as pd

df = pd.read_excel('/content/Data Ekspor Impor Nasional HS 2 Digit Agustus 2025.xlsx')
display(df.head(10))
```

```python
display(df.dtypes)
display(df.describe())
display(df.info())
```

```python
missing_values = df.isnull().sum()
display(missing_values)
```

```python
# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# If duplicate rows exist, remove them
if duplicate_rows > 0:
    df = df.drop_duplicates()

display(df.info())
```

```python
df.to_csv("data_ekspor_impor_bersih.csv", index=False)
print("Data telah berhasil disimpan ke dalam file data_ekspor_impor_bersih.csv")
```
